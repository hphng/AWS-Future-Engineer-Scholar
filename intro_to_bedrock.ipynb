```python
import boto3
import json
```

```python
session = boto3.Session(profile_name="udacity")
bedrock = session.client('bedrock', region_name='us-east-1')

response = bedrock.list_foundation_models()

modelIDs = [model['modelId'] for model in response['modelSummaries']]

print(json.dumps(modelIDs, indent=4))
```

```python
client = session.client('bedrock-runtime', region_name='us-east-1')
```

```python
model_id = "anthropic.claude-3-sonnet-20240229-v1:0"
```

```python
transcript = "story about an software engineer who tries to find job"
```

```python
prompt_text = f"""
<context>
You are a creative story teller. Given a text, you create a positive high energy story
</context>

<instruction>
Based on the following transcript, tell me a creative story that is positive and high energy, no more than 100 wordd.
</instruction>

<text>
{transcript}
</text>
"""
```

```python
request_body = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 1024,
    "temperature": 1,
    "top_p": 0.999,
    "messages": [
        {
            "role": "user",
            "content": [{"type": "text", "text": prompt_text}]
        }
    ]
}
```

```python
response = client.invoke_model(
    modelId=model_id,
    body=json.dumps(request_body)
)
```

```python
#parse response
response_body = json.loads(response['body'].read())
generated_text = response_body['content'][0]['text']
print("Generated Text:", generated_text)
```
