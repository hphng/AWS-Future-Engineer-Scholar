```python
import boto3
import json
```

```python
s3 = boto3.client("s3", region_name="us-east-1")
response = s3.get_object(
    Bucket="huy-udacity",
    Key='heros.txt'
)

heros = response['Body'].read().decode('utf-8')
# print(heros[:500])  

response = s3.get_object(
    Bucket="huy-udacity",
    Key='transcript.txt'
)

transcript = response['Body'].read().decode('utf-8')
```

```python
client = boto3.client('bedrock-runtime', region_name='us-east-1')
```

```python
model_id = "anthropic.claude-3-sonnet-20240229-v1:0"
```

```python
prompt_text = f"""
<context>
{heros}
</context>

<instruction>
You are to give me a bullet point summary of what happened during the transcript.
As well, please provide a sentiment analysis of the transcript.
Please provide your response in a valid JSON format with the following keys:
- summary: a bullet point summary of what happened during the transcript
- sentiment: a sentiment analysis of the transcript
</instruction>

<text>
{transcript}
</text>
"""
```

```python
request_body = {
    "anthropic_version": "bedrock-2023-05-31",
    "max_tokens": 1024,
    "temperature": 1,
    'top_p': 0.999,
    "messages": [
        {
            "role": "user",
            "content": [{"type": "text", "text": prompt_text}]
        }
    ]
}
```

```python
response = client.invoke_model(
    modelId=model_id,
    body=json.dumps(request_body)
)
```

```python
# Parse the response
response_body = json.loads(response['body'].read())
generated_text = response_body['content'][0]['text']
print(generated_text)
```
